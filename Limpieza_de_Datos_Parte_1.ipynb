```python
city_replacements = {
    'DUSSELDURF': 'DUSSELDORF',
    'DUESSELDORF': 'DUSSELDORF',
    'NUREMBERG': 'NURNBERG',
    'MUNICH': 'MUNCHEN',
    'KOLN': 'KOLN',
    'WARSAW': 'WARSZAWA',
    'CRACOW': 'KRAKOW',
    'CRACOVIA': 'KRAKOW',
    'ZURICH': 'ZURICH',
    'LÜBECK': 'LUEBECK',
    'LUTTICH': 'LIEGE',
    'HEIDELBERG': 'HEIDELBERG',
    'MANNHEIM': 'MANNHEIM',
    'STUTTGART': 'STUTTGART',
    'RIGA, LATVIA': 'RIGA',
    # Añadir más correcciones según sea necesario
}
```

```python
def clean_city(value):
    """
    Funcion para limpiar los datos en la columna "City"
    """
    if pd.isna(value):
        return value  # Si es NaN, lo dejamos como esta


    # Eliminar lo que está entre parentesis
    value = re.sub(r'\s*\(.*?\)', '', value)

    # Reemplazar 'NJ, USA' por 'New Jersey'
    if value == 'NJ, USA':
        value = 'New Jersey'

    value = value.lower()
    # Eliminar diéresis (ü, ö) y otros caracteres especiales
    value = value.replace('ü', 'u').replace('ö', 'o').replace('è', 'e')

    if value == 'Prefer not to say':
        return np.nan

    return value

def clean_and_standardize_city(value):
    """
    Función para limpiar los datos en la columna "City" y estandarizar las ciudades.
    """
    if pd.isna(value):
        return value  # Si es NaN, lo dejamos como está

    # Limpiar con la función clean_city
    value = clean_city(value)

    # Eliminar espacios al inicio y al final
    value = value.strip()

    # Convertir a mayúsculas (después de eliminar caracteres especiales)
    value = value.upper()

    # Reemplazar según el diccionario 'city_replacements'
    return city_replacements.get(value, value)  # Si está en el diccionario, reemplaza; si no, deja igual
```

```python
# Limpiar datos en la columna "City"
df['City'] = df['City'].apply(clean_and_standardize_city)
df['City'].unique()
```

# "Total years of experience"

```python
# Funcion posible para utilizar
def custom_round(value):
    """
    Funcion para redondear una columna a su decimal mas cercano
    """
    if pd.isna(value):
        return np.nan  # Dejamos NaN como está
    if isinstance(value, float):
        decimal_part = value - int(value)  # Extraemos la parte decimal
        if decimal_part <= 0.5:
            return np.floor(value)  # Redondea hacia abajo
        else:
            return np.ceil(value)   # Redondea hacia arriba
    else:
        return value
```

```python
# Verificar como estan escritos los datos
df['Total years of experience'].unique()
```

```python
df['Total years of experience'].describe()
```

```python
def clean_experience(value):
    """
    Funcion para convertir la columna "Total years of experience" a numerica
    """
    if pd.isna(value):
        return np.nan
    if value == 'less than year':
        return 0
    if value == '1 (as QA Engineer) / 11 in total':
        return 11
    if value == '15, thereof 8 as CTO':
        return 15
    if value == '6 (not as a data scientist, but as a lab scientist)':
        return 6
    if isinstance(value, str):
        value = value.replace(',', '.')
    try:
        return float(value)  # Intentar convertir a número
    except ValueError:
        return np.nan  # Si no es posible, devolver NaN
```

```python
# Limpiar datos y convertir columna "Total years of experience" a numerica
df['Total years of experience'] = df['Total years of experience'].apply(clean_experience)
df
```

```python
def split_other_technologies(value):
    """
    Función para dividir las tecnologías en columnas separadas.
    """
    if pd.isna(value):
        return pd.Series([np.nan, np.nan, np.nan, np.nan])

    # Convertir a lista y eliminar espacios innecesarios
    technologies = [tech.strip() for tech in value.split(',')]

    # Retornar los primeros cuatro valores o NaN si faltan tecnologías
    return pd.Series(technologies + [np.nan] * (4 - len(technologies)))[:4]
```

```python
def clean_vacation_days(value):
    """
    Función para limpiar la columna "Number of vacation days"
    y convertirla a numérica
    """
    if pd.isna(value):
        return np.nan

    # Eliminar caracteres no deseados y convertir a número
    if isinstance(value, str):
        value = value.strip()  # Eliminar espacios en blanco iniciales y finales
        value = value.lower() # Convertir a minusculas
        if value in ['unlimited']:
            return np.nan
        elif value == '(no idea)':
            return np.nan
        elif value == '30 in contract (but theoretically unlimited)':
            return 30
        elif value == '24 labour days':
            return 24
        elif value == '~25':
            return 25
        elif value == '23+':
            return 23
        else:
            # Intentar convertir a número si es posible
            try:
                return float(value)
            except ValueError:
                return np.nan
    else:
        return np.nan
```

# "Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR"

```python
df['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'].unique()
```

```python
def clean_value(value):
    if isinstance(value, str):
        # Eliminar texto no numérico irrelevante y símbolos no deseados
        value = value.lower()  # Convertir todo a minúsculas para evitar duplicados como 'No' y 'no'

        # Filtrar valores claramente irrelevantes
        if any(keyword in value for keyword in ['no', 'yes', '-', 'voucher', 'equipment']):
            return np.nan

        # Remover símbolos de moneda como €, eliminar comas y espacios extra
        value = re.sub(r'[€,\s]', '', value)

        # Manejo de porcentajes, puedes decidir si lo conviertes o lo eliminamos
        if '%' in value:
            return np.nan

        # Remover texto no deseado que acompaña a números (como "per month" o similar)
        value = re.sub(r'[a-zA-Z]', '', value)

        # Intentar convertir a número
        try:
            return float(value)
        except ValueError:
            return np.nan
    return value
```

```python
df['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'] = df['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'].apply(clean_value)
df['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'] = df['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'].astype(float)
```

```python
df.rename(columns={'Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR':'additional monetary support'}, inplace = True)
```
