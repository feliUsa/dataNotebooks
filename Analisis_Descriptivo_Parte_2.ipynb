## AE Visualizar serie de tiempo de ganancias

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Rango de fechas
df_filtered = df[(df['Timestamp'] >= '2020-11-24') & (df['Timestamp'] <= '2021-01-19')]
```

```python
# Calcular el IQR (rango intercuartílico)
Q1 = df_filtered['Yearly brutto salary in thousands EUR'].quantile(0.25)
Q3 = df_filtered['Yearly brutto salary in thousands EUR'].quantile(0.75)
IQR = Q3 - Q1

# Rango
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar temporalmente los valores fuera de este rango para la visualización
df_no_outliers = df_filtered[
    (df_filtered['Yearly brutto salary in thousands EUR'] >= lower_bound) &
    (df_filtered['Yearly brutto salary in thousands EUR'] <= upper_bound)
]
```

```python
# Serie de tiempo sin valores atipicos
plt.figure(figsize=(10, 6))
plt.plot(df_no_outliers['Timestamp'], df_no_outliers['Yearly brutto salary in thousands EUR'], marker='o', color='orchid')
plt.title('Yearly Brutto Salary in Thousands EUR (2020-11-24 to 2021-01-19)')
plt.xlabel('Date')
plt.ylabel('Yearly Brutto Salary in Thousands EUR')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
all_technologies = []
salaries = []

for index, row in df.iterrows():
    salary = row['Yearly brutto salary in thousands EUR']
    for column in ['Main technology', 'Other technologies', 'Other technologies 2', 'Other technologies 3', 'Other technologies Extra']:
        tech = row[column]
        if isinstance(tech, str):
            for t in tech.split(', '):
                all_technologies.append(t.strip())
                salaries.append(salary)  # Append the salary for this technology

# Create a DataFrame with technology and salary
technology_salary_df = pd.DataFrame({'Technology': all_technologies, 'Salary': salaries})

# Group by technology and calculate average salary
technology_avg_salary = technology_salary_df.groupby('Technology')['Salary'].agg(['mean', 'max']).sort_values(by=['mean', 'max'], ascending=False)

# Visualize the average salary for each technology
plt.figure(figsize=(15, 8))
technology_avg_salary['mean'].plot(kind='bar', color='darkmagenta')
plt.title('Salario promedio por tecnologia')
plt.xlabel('Tecnologia')
plt.ylabel('Salario promedio (Thousands EUR)')
plt.xticks(rotation=90)
plt.ylim(0, 180)
plt.tight_layout()
plt.show()
```
